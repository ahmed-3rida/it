# 1️⃣ المنطق الأساسي للمنصة (`shared.js`)
هذا الملف هو "المحرك" الذي يعمل خلف الكواليس في كل صفحات الموقع.

### 🌐 نظام الترجمة الذكي (`applyLanguage`)
نستخدم نظاماً يعتمد على السمات (Attributes) لترجمة العناصر دون الحاجة لملفات ترجمة معقدة.

**شرح الكود:**
1. نتحقق من اللغة المختارة (`ar` أو `en`).
2. نغير اتجاه الموقع `document.documentElement.dir` (rtl للعربي و ltr للإنجليزي).
3. نستخدم `querySelectorAll('[data-ar]')` للوصول لجميع العناصر القابلة للترجمة.
4. نقوم بتحديث الـ `innerText` بناءً على قيمة السمة الموافقة للغة.

```python
// وظيفة الترجمة الأساسية
function applyLanguage(lang) {
    // تحديث اتجاه الصفحة
    document.documentElement.dir = (lang === 'en') ? 'ltr' : 'rtl';

    // ترجمة النصوص
    document.querySelectorAll('[data-ar]').forEach(el => {
        el.innerText = (lang === 'en') ? el.getAttribute('data-en') : el.getAttribute('data-ar');
    });
}
```

# 2️⃣ نظام الدخول والتسجيل (`login_register/`)
المسؤول عن إدارة جلسات المستخدم وحفظ البيانات محلياً.

### 🔐 وظيفة تسجيل الدخول (`doLogin`)
هذه الوظيفة لا تتواصل مع سيرفر، بل تستخدم تقنية `localStorage` لمحاكاة قاعدة البيانات.

**ماذا يحدث عند الضغط على "دخول"؟**
1. يتم تخزين قيمة `isLoggedIn = true` (لكي يعرف الموقع في المرة القادمة أنك مسجل).
2. يتم تخزين الـ `username` لكي نظهره في الـ Navbar والمنشورات.
3. يتم تحويل المستخدم لصفحة الـ `home` باستخدام `window.location.href`.

```python
function doLogin(username) {
    localStorage.setItem('isLoggedIn', 'true');
    localStorage.setItem('username', username || 'مستخدم');
    window.location.href = '../home/index.html';
}
```

# 3️⃣ الصفحة الرئيسية والتغذية الإخبارية (`home/`)
الجزء الأكثر تعقيداً في المشروع، حيث يتم بناء المحتوى ديناميكياً.

### 🧬 دمج وترتيب البيانات (`Combined Feed`)
لجعل التغذية الإخبارية متنوعة، نقوم بدمج المنشورات مع الاستطلاعات.

**المنطق المتبع:**
1. جلب المصفوفات من الـ `localStorage`.
2. استخدام الـ Spread Operator `[...]` لدمجهم.
3. الترتيب باستخدام `sort()` بناءً على الوقت (ID) لضمان ظهور الأحدث في الأعلى.

```python
let myPosts = JSON.parse(localStorage.getItem('myPosts') || '[]');
let customPolls = JSON.parse(localStorage.getItem('customPolls') || '[]');

// دمج المصفوفات
let combinedFeed = [...myPosts, ...customPolls];

// الترتيب من الأحدث للأقدم
combinedFeed.sort((a, b) => b.id - a.id);
```

# 💡 نصائح ذهبية للمناقشة (Defense Tips)

1. **سؤال: لماذا استخدمت `localStorage` وليس قاعدة بيانات حقيقية؟**
   * الإجابة: لأن المشروع يركز على تقنيات الـ Front-end و Vanilla JS، والـ LocalStorage توفر سرعة هائلة وتجربة مستخدم ممتازة دون الحاجة لإعدادات سيرفر معقدة.

2. **سؤال: كيف تضمن أن الموقع يعمل باللغتين في كل الصفحات؟**
   * الإجابة: من خلال تضمين ملف `shared.js` في كل الصفحات واستدعاء وظيفة `applyLanguage()` فور تحميل الصفحة (DOMContentLoaded).

3. **سؤال: ما هو أصعب جزء واجهته؟**
   * الإجابة: دمج أنواع مختلفة من البيانات (منشورات واستطلاعات) في قائمة واحدة وترتيبها زمنياً مع الحفاظ على الأداء.
